```python
import turtle
v = turtle.Turtle()
def hinh1(a,b):
        v.fillcolor('red')
        v.begin_fill()
        for i in range(1, 3):
            v.forward(a)
            v.left(90)
            v.forward(b)
            v.left(90)
        v.end_fill()
def hinh2(a,b,c,d):
            v.penup()
            v.goto(c,d)
            v.pendown()
            v.fillcolor('red')
            v.begin_fill()
            for i in range(1, 3):
                v.forward(a)
                v.left(90)
                v.forward(b)
                v.left(90)
            v.end_fill()
def cua_so(c):
        v.fillcolor('white')
        v.begin_fill()
        for a in range(1,c):
            v.left(55)
            v.penup()
            v.forward(75)
            v.pendown()
            v.right(55)
        for j in range(1,c+1):
            v.forward(60)
            v.left(90)
            v.forward(100)
            v.left(90)
        v.end_fill()
def vuong(d,f,g):
            v.penup()
            v.goto(300,0)
            v.pendown()
            v.fillcolor('blue')
            v.begin_fill()
            for i in range(1, 3):
                v.forward(f)
                v.left(90)
                v.forward(d)
                v.left(90)
            v.end_fill()
            v.forward(175)
            v.fillcolor('green')
            v.begin_fill()
            for j in range(1, 3):
                v.left(90)
                v.forward(g)
                v.left(90)
                v.forward(g)
            v.end_fill()
            v.back(50)
            v.left(90)
            v.forward(g)
            turtle.done()
hinh1(150,200)  
cua_so(2)
hinh2(150,200,0,200)  
cua_so(2)
hinh2(150,200,150,0) 
cua_so(2)
hinh2(150,200,150,200) 
cua_so(2)
vuong(200,250,100)
```
