```python

danh_sach_sp = {}

def them_san_pham(danh_sach_sp):
    print('== Chức năng thêm sản phẩm ==')
    while True:
        print("= Nếu muốn thoát chức năng thêm sản phẩm vui lòng ấn 'y' để thoát. ")
        print("= Nếu tiếp tục ấn phím bất kì. ")
        lua_chon = input("nhập lựa chọn: ")
        if lua_chon == 'y':
            return
        ma_sp = input('Nhập mã sản phẩm: ')
        ten_sp = input('Nhập tên sản phẩm: ')
        gia_sp = input('Nhập giá sản phẩm: ')
        if ma_sp in danh_sach_sp:
            print(f'Sản phẩm có mã là {ma_sp} đã tồn tại vui lòng nhập lại.')
        else:
            danh_sach_sp[ma_sp] = {
                'Tên': ten_sp,
                'Giá Tiền': gia_sp
            }
            print(f'Thêm thành công thông tin sản phẩm với mã là {ma_sp}.')


def sua_san_pham(danh_sach_sp):
    print('== Chức năng sửa sản phẩm ==')
    while True:
        print("= Nếu muốn thoát chức năng sửa sản phẩm vui lòng ấn 'y' để thoát. ")
        print("= Nếu tiếp tục ấn phím bất kì. ")
        lua_chon = input("nhập lựa chọn: ")
        if lua_chon == 'y':
            return
        ma_sp = input('Nhập mã sản phẩm cần sửa: ')
        if ma_sp not in danh_sach_sp:
            print('Mã sản phẩm không tồn tại. Vui lòng nhập lại mã.')
        else:
            ten_sp = input('Nhập tên mới cho sản phẩm: ')
            gia_sp = input('Nhập giá sản phẩm: ')
            danh_sach_sp[ma_sp] = {
                'Tên': ten_sp,
                'Giá Tiền': gia_sp
            }
            print(f'Cập nhật thành công thông tin sản phẩm có mã là {ma_sp}.')

def xoa_san_pham(danh_sach_sp):
    print('== Chức năng xóa sản phẩm ==')
    while True:
        print("= Nếu muốn thoát chức năng sửa sản phẩm vui lòng ấn 'y' để thoát. ")
        print("= Nếu tiếp tục ấn phím bất kì. ")
        lua_chon = input("nhập lựa chọn: ")
        if lua_chon == 'y':
                return
        ma_sp = input('Nhập mã sản phẩm cần xóa: ')
        if ma_sp not in danh_sach_sp:
            print('Mã sản phẩm không tồn tại. Vui lòng nhập lại mã.')
        else:
            danh_sach_sp.pop(ma_sp)
            print(f'Xóa thành công mã sản phẩm là {ma_sp}.')

def hien_thi_danh_sach_sp(danh_sach_sp):
    print('== Chức năng hiện thị thông tin danh sách sản phẩm ==')
    if not danh_sach_sp:
        print("Danh sách sản phẩm hiện tại trống. Vui lòng thêm sản phẩm.")
    for ma_sp, thong_tin in danh_sach_sp.items(): 
        print(f'Mã: {ma_sp}. Tên: {thong_tin["Tên"]}. Giá Tiền: {thong_tin["Giá Tiền"]}')

while True:
    print('=== Chương Trình quản lý danh sách sản phẩm ===')
    print('Chọn thao tác:')
    print('1. Chức năng thêm sản phẩm.')
    print('2. Chức năng sửa sản phẩm.')
    print('3. Chức năng xóa sản phẩm.')
    print('4. Chức năng hiện thị thông tin danh sách sản phẩm.')
    print('5. Thoát chương trình.')
    lua_chon = input('Vui lòng nhập thao tác (1-5): ')
    if lua_chon == '1':
        them_san_pham(danh_sach_sp)
    elif lua_chon == '2':
        sua_san_pham(danh_sach_sp)
    elif lua_chon == '3':
        xoa_san_pham(danh_sach_sp)
    elif lua_chon == '4':
        hien_thi_danh_sach_sp(danh_sach_sp)
    elif lua_chon == '5':
        print("Cảm ơn quý khách sử dụng sản phẩm của chúng tôi.")
        break
    else:
        print('Thao tác không tồn tại. Vui lòng nhập lại.')







```
