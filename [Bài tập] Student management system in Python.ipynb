```python
class ThongTinSv:
    def __init__(self, ho_ten ,ma_sv , ntns, que_quan, lop, chuyen_nganh):
        self.ten = ho_ten
        self.ma_sv = ma_sv
        self.ntns = ntns
        self.que = que_quan
        self.lop = lop
        self.nganh = chuyen_nganh
    def get_ma_sv(self):
        return self.ma_sv
class ChucNang:
    def __init__(self):
        self.danh_sach = []
    def hien_thi_ds(self):
        print('== Chức năng hiện thị thông tin danh sách sinh viên ==')
        if not self.danh_sach:
            print('Hiện tại danh sách đang trống xin mời nhập thêm thông tin sinh viên vào danh sách.')
        else:
            print('Mã Sinh Viên\t\t Họ&Tên\t\t Năm Sinh\t\t Quên Quán\t\t Lớp\t\t Chuyên Ngành')
            for thong_tin in self.danh_sach:
                print(f'{thong_tin.ma_sv:<22} {thong_tin.ten:<16} {thong_tin.ntns:<23} {thong_tin.que:<23} {thong_tin.lop:<20} {thong_tin.nganh}')
    def them_ttsv(self):
        print('=== Chức năng thêm thông tin sinh viên ===')
        while True:
            print("= Nếu muốn thoát chức năng thêm sản phẩm vui lòng ấn 'y' để thoát. ")
            print("= Nếu tiếp tục ấn phím bất kì. ")
            lua_chon = input("nhập lựa chọn: ")
            if lua_chon == 'y':
                return
            ma_sv = input('Nhập mã sinh viên : ')
            ho_ten = input('Nhập họ và tên sinh viên: ')
            ntns = input('Nhập ngày tháng năm sinh của sinh viên: ')
            que_quan = input('Nhập quê quán của sinh viên: ')
            lop = input('Nhập tên lớp của sinh viên: ')
            chuyen_nganh = input('Nhập chuyển ngành của sinh viên: ')
            thong_tin = ThongTinSv(ho_ten ,ma_sv , ntns, que_quan, lop, chuyen_nganh)
            self.danh_sach.append(thong_tin)
            print('*** Thêm thành công thông tin sinh viên ***')
    def sua_thong_tin(self):
        print('=== Chức năng sửa thông tin sinh viên ===')
        while True:
            print("= Nếu muốn thoát chức năng thêm sản phẩm vui lòng ấn 'y' để thoát. ")
            print("= Nếu tiếp tục ấn phím bất kì. ")
            lua_chon = input("nhập lựa chọn: ")
            if lua_chon == 'y':
                return
            ma_sv = input("Nhập mã sinh viên cần sửa: ")
            print('Nếu không thay đổi thông tin thì thông tin cũ vẫn giữ lại.')
            for i in self.danh_sach:
                if i.get_ma_sv() == ma_sv:
                    i.ten = input('Nhập họ và tên sinh viên: ')
                    i.ntns = input('Nhập ngày tháng năm sinh của sinh viên: ')
                    i.que = input('Nhập quê quán của sinh viên: ')
                    i.lop = input('Nhập tên lớp của sinh viên: ')
                    i.nganh = input('Nhập chuyển ngành của sinh viên: ')
                    print(f"Sửa thành công thông tin có mã sinh viên là {ma_sv}.")
            if ma_sv not in self.danh_sach:
                print(f'*** Trong danh sách sinh viên không có mã sinh viên: {ma_sv}. ***')
                print('*** Xin thêm thông tin sinh viên vào danh sách. ***')
    def xóa_thong_tin(self):
        print('=== Chức năng xóa thông tin sinh viên ===')
        while True:
            print("= Nếu muốn thoát chức năng thêm sản phẩm vui lòng ấn 'y' để thoát. ")
            print("= Nếu tiếp tục ấn phím bất kì. ")
            lua_chon = input("nhập lựa chọn: ")
            if lua_chon == 'y':
                return
            ma_sv = input("Nhập mã sinh viên cần xóa: ")
            for i in self.danh_sach:
                if i.get_ma_sv() == ma_sv:
                    self.danh_sach.remove(i)
                    print(f'Đã xóa thành công thông tin sinh viên có mã là {ma_sv}.')
            if ma_sv not in self.danh_sach:
                print(f'*** Trong danh sách sinh viên không có mã sinh viên: {ma_sv} ***.')
                print('*** Xin thêm thông tin sinh viên vào danh sách. ***')
    def tim_kiem(self):
        print('=== Chức năng tìm kiếm thông sinh viên ===')
        while True:
            print("= Nếu muốn thoát chức năng thêm sản phẩm vui lòng ấn 'y' để thoát. ")
            print("= Nếu tiếp tục ấn phím bất kì. ")
            lua_chon = input("nhập lựa chọn: ")
            if lua_chon == 'y':
                return
            ma_sv = input("Nhập mã sinh viên cần tìm: ")
            for i in self.danh_sach:
                if i.get_ma_sv() == ma_sv:
                    print('Mã Sinh Viên\t\t Họ&Tên\t\t Năm Sinh\t\t Quên Quán\t\t Lớp\t\t Chuyên Ngành')
                    print(f'{i.ma_sv}\t {i.ten}\t {i.ntns}\t {i.que}\t {i.lop}\t {i.nganh}')
    def sap_xep_tt(self):
        def sap_xep(sinh_vien):
            tam = sinh_vien.ten.split()
            return tam[-1]
        print('=== Chức năng sắp xếp danh sách sinh viên ===')
        self.danh_sach.sort(reverse= False,key=sap_xep )
start = ChucNang()
while True:
    print('=== Chương trình quản lý sinh viên === \n\
    1.Xem danh mục sinh viên. \n\
    2.Thêm sinh viên mới vào danh mục. \n\
    3.Cập nhật thông tin của sinh viên trong danh mục. \n\
    4.Xóa thông tin sinh viên khỏi danh mục. \n\
    5.Tìm kiếm thông tin sinh viên trong danh mục theo từ khóa \n\
    6.Sắp xếp thông tin sinh viên trong danh mục. \n\
    7.Thoát chương trình.')
    thao_tac = input('Xin chọn thao tác chức năng của chương trình (1-7): ')
    if thao_tac == '1':
        start.hien_thi_ds()
    elif thao_tac == '2':
        start.them_ttsv()
    elif thao_tac == '3':
        start.sua_thong_tin()
    elif thao_tac == '4':
        start.xóa_thong_tin()
    elif thao_tac == '5':
        start.tim_kiem()
    elif thao_tac == '6':
        start.sap_xep_tt()
    elif thao_tac == '7':
        print("*** Cảm ơn bạn đã dùng chuong trình. See you again. ***")
        break
    else:
        print('Nhập sai thao tác. Xin mời nhập lại.')
```
