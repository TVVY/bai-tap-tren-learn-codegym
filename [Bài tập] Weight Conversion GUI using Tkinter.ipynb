```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

win = tk.Tk()
win.title('Chuyển đổi cân nặng')
win.geometry('800x150')

def convert_click():
    kg = kg_input.get()
    try:
        gram = 1000 * (float(kg))
        pounds = 2.20462 * (float(kg))
        ounce = 35.274 * (float(kg))
    except:
        messagebox.showerror(title='Error',message='lỗi nhập chữ, xin mời nhập bằng số')
        return
    gram_output.config(text= gram)
    pounds_output.config(text= pounds)
    ounce_output.config(text= ounce)
#1
labell = tk.Label(
    text = 'Enter the weigth in KG ',
    font=("arial",17,"normal")
)
labell.grid(sticky=tk.E)
#get
kg_input = tk.Entry(
    font=("arial",17,"normal")
)
kg_input.grid(column=1, row=0)
#convert
convert = tk.Button(
    text = 'convert',
    font=("arial",17,"normal"),
    command = convert_click
)
convert.grid(column=2, row=0)

#gram
gram = tk.Label(
    text = 'Gram',
    font=("arial",17,"normal")
)
gram.grid(column=0, row=1,padx=100)
#Pounds
Pounds = tk.Label(
    text = 'Pounds',
    font=("arial",17,"normal")
)
Pounds.grid(column=1, row=1,padx=100)
#Ounce
Ounce = tk.Label(
    text = 'Ounce',
    font=("arial",17,"normal")
)
Ounce.grid(column=2, row=1,padx=100)
#row
gram_output = tk.Label(text='0',font=("arial",17,"normal"))
gram_output.grid(column=0, row=3)
pounds_output = tk.Label(text='0',font=("arial",17,"normal"))
pounds_output.grid(column=1,row=3)
ounce_output = tk.Label(text='0',font=("arial",17,"normal"))
ounce_output.grid(column=2,row=3)
win.mainloop()
```
