```python
danh_sach_chi_tieu = {}
def them_chi_tieu(danh_sach_chi_tieu):
    print('== Chức năng thêm chi tiêu ==')
    while True:
        print("= Nếu muốn thoát chức năng thêm chi tiêu vui lòng ấn 'y' để thoát. ")
        print("= Nếu tiếp tục ấn phím bất kì. ")
        chon = input("nhập lựa chọn: ")
        if chon == 'y':
            return
        else:
            ten_chi_tieu = input("Nhập tên chi tiêu: ")
            tien_chi = input("Nhập số tiền chi tiêu: ")
            ngay_chi_tien = input("Nhập ngày chi tiêu: ")
            danh_sach_chi_tieu[ten_chi_tieu] = {
                'Số tiền chi' : tien_chi,
                'Ngày chi' : ngay_chi_tien
            }
            print("= Thêm thành công chi tiêu =")
def hien_thi_chi_tieu(danh_sach_chi_tieu):
    print('== Chức năng hiện thị thông tin danh sách sản phẩm ==')
    if not danh_sach_chi_tieu:
        print("Danh sách chi tiêu hiện tại trống. Vui lòng thêm thông tin chi tiêu vào danh sách.")
    for ten_chi_tieu, thong_tin in danh_sach_chi_tieu.items():
        print(f'Tên: {ten_chi_tieu}. Số tiền chi: {thong_tin["Số tiền chi"]}. Ngày chi: {thong_tin["Ngày chi"]}')

def xoa_chi_tieu(danh_sach_chi_tieu):
    print('== Chức năng xóa chi tiêu ==')
    while True:
        print("= Nếu muốn thoát chức năng xóa chi tiêu vui lòng ấn 'y' để thoát. ")
        print("= Nếu tiếp tục ấn phím bất kì. ")
        lua_chon = input("nhập lựa chọn: ")
        if lua_chon == 'y':
            return
        else:
            ten_chi_tieu = input("Nhập tên chi tiêu cần xóa.")
            if ten_chi_tieu not in danh_sach_chi_tieu:
                print("Tên chi tiêu không có trong danh sách xin mời nhập lại.")
            else:
                danh_sach_chi_tieu.pop(ten_chi_tieu)
                print(f'Xóa thành công tên chi tiêu: {ten_chi_tieu}.')
while True:
    print("Quý khách muốn thực hiện thao tác nào? -\n"\
            "1. Thêm chi tiêu.\n" \
            "2. Xóa chi tiêu.\n" \
            "3. Hiện thị danh sách chi tiêu.\n"\
            "4. Thoát ứng dụng chi tiêu.")
    lua_chon = input("== Xin mời nhập lựa chọn (1-4): ")
    if lua_chon == '1':
        them_chi_tieu(danh_sach_chi_tieu)
    elif lua_chon == '2':
        xoa_chi_tieu(danh_sach_chi_tieu)
    elif lua_chon == '3':
        hien_thi_chi_tieu(danh_sach_chi_tieu)
    elif lua_chon == '4':
        print("=== Cảm ơn Quý Khách đã sử dụng sản phẩm ===")
        break
    else:
        print("=== Thao tác không đúng xin mời nhập lại ===")
```
