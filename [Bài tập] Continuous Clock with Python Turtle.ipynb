```python
import turtle
import datetime
import math
import time

screen = turtle.Screen()
screen.title('[Bài tập-Trên Lear] Vẽ Đồng Hồ-CodeGym By TẠ VĂN VỸ')
screen.bgcolor('sky blue')
screen.setup(width= 1000, height= 1000)
screen.setworldcoordinates(-1000,-1000,1000,1000)
screen.tracer(0)


class clock:
    def __init__(self,hour,minute,second):
        self.hour, self.minute, self.second = hour, minute, second
        self.microsecond = 0
        self.mat = turtle.Turtle()
        self.kim = turtle.Turtle()
        self.mat.hideturtle()
        self.kim.hideturtle()

    def ve(self):
        self.ve_mat()
        self.ve_kim()
        
    def ve_mat(self):
        #vẽ mặt đồng hồ
        self.mat.clear()
        self.mat.up()
        self.mat.goto(0,-700)
        self.mat.pensize(4)
        self.mat.down()
        self.mat.fillcolor('white')
        self.mat.begin_fill()
        self.mat.circle(700,steps=100)
        self.mat.end_fill()
        self.mat.up()
        self.mat.goto(0,0)
        self.mat.dot(10)
        self.mat.pensize(2)
        for ve_phut in range(0,360,6):
            self.mat.up()
            self.mat.goto(0,0)
            self.mat.seth(90-ve_phut)
            self.mat.fd(620)
            self.mat.down()
            self.mat.fd(30)
            
        self.mat.pensize(3)
        for angle in range(0,360,30):
            self.mat.up()
            self.mat.goto(0,0)
            self.mat.seth(90-angle)
            self.mat.fd(600)
            self.mat.down()
            self.mat.fd(50)
            
        self.mat.pensize(6)
        for angle in range(0,360,90):
            self.mat.up()
            self.mat.goto(0,0)
            self.mat.seth(90-angle)
            self.mat.fd(580)
            self.mat.down()
            self.mat.fd(70)
        
    def Ve_KDH(self):
        #Ve_KDH là vẽ kim đồng hồ
        self.kim.clear()       
        self.kim.up()
        self.kim.goto(0,0)
        self.kim.seth(90-math.floor(((self.hour%12)*60*60*1000000+self.minute*60*1000000+self.second*1000000+self.microsecond)/3600000000*30))
        self.kim.down()
        self.kim.color('blue')
        self.kim.pensize(6)
        self.kim.fd(300)

        self.kim.up()
        self.kim.goto(0,0)
        self.kim.seth(90-math.floor((self.minute*60*1000000+self.second*1000000+self.microsecond)/60000000*6))
        self.kim.down()
        self.kim.color('black')
        self.kim.pensize(4)
        self.kim.fd(400)

        self.kim.up()
        self.kim.color('red')
        self.kim.goto(0,0)
        self.kim.dot(5)
        self.kim.seth(90-(self.second*1000000+self.microsecond)/1000000*6)
        self.kim.down()
        self.kim.pensize(2)
        self.kim.fd(570)

def Time():
    #setup thời gian cho đồng hồ
    global c
    d = datetime.datetime.now()
    c.hour, c.minute, c.second, c.microsecond = d.hour, d.minute, d.second, d.microsecond
    c.Ve_KDH()
    screen.update()
    screen.ontimer(Time,100)
def Year():
    v = turtle.Turtle()
    year = datetime.datetime.now().year
    month = datetime.datetime.now().month
    day = datetime.datetime.now().day
    v.hideturtle()
    v.up()
    v.goto(-250,850)
    v.write(str(day).zfill(2)
        +"/"+str(month).zfill(2)+"/"
        +str(year).zfill(2),
        font =("Arial", 35, "bold"))
    v.goto(0,0)
    v.down
def gio():
    v = turtle.Turtle()
    second = datetime.datetime.now().second
    minute = datetime.datetime.now().minute
    hour = datetime.datetime.now().hour
    v.hideturtle()
    v.up()
    v.goto(-200,750)
    while True:
        v.clear()
        v.write(str(hour).zfill(2)
                +":"+str(minute).zfill(2)+":"
                +str(second).zfill(2),
                font =("Arial", 35, "bold"))
        time.sleep(1)
        second += 1
        if second == 60:
            second = 0
            minute+= 1

        if minute == 60:
            minute = 0
            hour+= 1

        if hour == 13:
            hour = 1
    v.down

d = datetime.datetime.now()
c = clock(d.hour,d.minute,d.second)
c.ve_mat()
screen.update()
Time()
Year()
gio()
screen.mainloop()
```
