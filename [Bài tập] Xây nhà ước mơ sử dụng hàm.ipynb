```python
import turtle
import math

def ve_hinh_chu_nhat(v, rong, cao):
    for i in range(0, 2):
        v.forward(rong)
        v.left(90)
        v.forward(cao)
        v.left(90)
def ve_tam_giac_can(v, canh_day, mau):
    canh_ben = canh_day / math.sqrt(2)
    v.fillcolor(mau)
    v.begin_fill()
    v.forward(canh_day)
    v.left(135)
    v.forward(canh_ben)
    v.left(90)
    v.forward(canh_ben)
    v.left(135)
    v.end_fill()
    return math.sqrt(pow(canh_ben, 2) - pow(canh_day/2, 2))
           
def ve_hinh_thang_can(v, canh_day, canh_tren, chieu_cao, mau):
    canh_ben = math.sqrt(pow((canh_day - canh_tren)/2, 2) + pow(chieu_cao, 2))
    tam =  (pow((canh_day - canh_tren)/2, 2) + pow(canh_ben, 2) - pow(chieu_cao, 2))  \
           / ((2 * canh_ben) * ((canh_day - canh_tren)/2))
    goc_ke_day = math.degrees(math.acos(tam))
    v.fillcolor(mau)
    v.begin_fill()
    v.forward(canh_day)
    v.left(180 - goc_ke_day)
    v.forward(canh_ben)
    v.left(goc_ke_day)
    v.forward(canh_tren)
    v.left(goc_ke_day)
    v.forward(canh_ben)
    v.end_fill()
def ve_cay(v, canh_day, toa_do_x, toa_do_y, mau):          
    v.penup()
    v.setheading(0)
    v.goto(toa_do_x, toa_do_y)
    v.pencolor(mau)
    v.pendown()
    ve_hinh_thang_can(v, canh_day/4, canh_day/6, 0.6*canh_day, 'brown')
    v.setheading(0)
    v.penup()
    v.forward((canh_day/4)/2)
    v.setheading(90)
    v.forward((0.6*canh_day))
    v.setheading(180)
    v.forward(canh_day/2)
    for i in range(0, 3):
        phan_tram_giam = 20 if i != 0 else 0
        canh_day = canh_day * (1- (phan_tram_giam/100))
        v.setheading(0)
        v.forward(canh_day * (phan_tram_giam/2/100))
        v.pendown()
        duong_cao = ve_tam_giac_can(v, canh_day, mau)
        v.penup()
        v.setheading(90)
        v.forward(duong_cao)
def ve_nha(v, rong, cao, toa_do_x, toa_do_y):
    #than nha va cua nha
    v.penup()
    v.goto(toa_do_x, toa_do_y)
    v.pendown()
    v.fillcolor('darkturquoise')
    v.begin_fill()
    v.pensize(2)
    ve_hinh_chu_nhat(v, rong, cao)
    v.end_fill()
    v.fillcolor('lightsalmon')
    v.begin_fill()
    v.forward(rong/2)
    v.backward(rong/5/2)
    ve_hinh_chu_nhat(v, rong/5, (3/4)*cao)
    v.end_fill()
    #ong khoi
    v.penup()
    v.goto(toa_do_x, toa_do_y)
    v.setheading(90)
    v.forward(cao)
    v.setheading(0)
    toa_do_bat_dau_mai = v.position()
    v.pendown()
    v.fillcolor('black')
    v.begin_fill()
    day_mai = (3/4)*rong
    v.forward(0.7*day_mai)
    ve_hinh_chu_nhat(v, 0.1*day_mai, 0.45*day_mai)
    v.end_fill()
    v.penup()                  
    v.forward(0.1*day_mai)                  
    v.setheading(90)                  
    v.forward(0.55*day_mai)
    v.pendown()
    v.forward(0.15*day_mai)
    v.penup()
    v.setheading(180)
    v.forward(0.1*day_mai)
    v.setheading(270)
    v.forward(((0.15*day_mai)/2)/2)
    v.pendown()
    v.forward((0.15*day_mai)/2)
#   mai
    v.penup()
    v.goto(toa_do_bat_dau_mai)
    v.setheading(0)
    v.pendown()
    ve_tam_giac_can(v, day_mai, 'darkgray')
    
    
    
    
v = turtle.Turtle()
turtle.screensize(1000, 800)
v.speed(0)
# ve nha
ve_nha(v, 300, 140, -200, -100,)
# ve cay nho
ve_cay(v, 100, -270, -180, 'green')
# ve cay lon    
ve_cay(v, 200, 250, -200, 'green') 

turtle.done()
```
